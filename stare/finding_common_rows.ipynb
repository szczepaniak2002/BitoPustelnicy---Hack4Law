```python
import pandas as pd
import random
random.seed(35)

df = pd.DataFrame(columns=['Art. ' + str(c) + '.' for c in range(1, 15)])

random_rows = []
for i in range(100):
    random_row = [random.choices([0,1], [0.5, 0.5])[0] for _ in range(14)]
    random_rows.append(random_row)

df = pd.concat([df, pd.DataFrame(random_rows, columns=df.columns)], ignore_index=True)

def common_ones(row1, row2):
    ones1 = set([i+1 for i, value in enumerate(row1) if value == 1])
    ones2 = set([i+1 for i, value in enumerate(row2) if value == 1])
    return ones1.intersection(ones2)
    

given_row = [random.choices([0,1], [0.5, 0.5])[0] for _ in range(14)]

df['intersection'] = df.apply(lambda row: common_ones(given_row, row), axis=1)
df['intersection_len'] = df.apply(lambda row: len(common_ones(given_row, row)), axis=1)

sorted_df = df.sort_values('intersection_len', ascending=False)
most_similar_rows = sorted_df.iloc[0:5]
df = df.drop('intersection', axis=1)
df = df.drop('intersection_len', axis=1)

display(most_similar_rows)
display(given_row)
```
